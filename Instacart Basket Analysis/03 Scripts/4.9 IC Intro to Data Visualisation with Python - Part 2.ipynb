### Sampling Data

```python
#creating a sample of the data by creating a list
np.random.seed(4)
dev = np.random.rand(len(all_data_merge)) <= 0.7
```

```python
dev
```

```python
#store the 70% of the sample to one df
big = all_data_merge[dev]
```

```python
#store the 30% of the sample to another df
small = all_data_merge[~dev]
```

```python
len(big)+len(small)
```

```python
#create a smaller subset of the data by only including the necessary columns
df_2 = small[['order_day_of_week','prices']]
```

```python
#create the line chart based on the subset data
line = sns.lineplot(data = df_2, x = 'order_day_of_week',y = 'prices', color ='#900C3F')
```

```python
#download the file
line.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'line_prices_dow.png'))
```

### Task

```python
#creating a histogram of hours of the day
hist_2 = small['order_hour_of_day'].plot.hist(bins = 70, color ='#ff5733')
plt.xlabel('Hour of the Day')
plt.ylabel('Frequency')
plt.title('Distribution of Orders by Hour of the Day')
```

```python
#export visualisation
hist_2.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'hist_orders_hod.png'))
```

From the distribution it shows that most customers place their orders between 09:00 and 16:00. Interesting is that orders are made throughout the day even at night, past midnight.

```python
#bar chart showing the cutomer loyalty numbers
bar_2 = small['loyalty_flag'].value_counts().sort_index().plot.barh(color =['#900C3F'])
plt.xlabel('Customers')
plt.ylabel('Loyalty Status')
plt.title('Customer Loyalty Status')
```

```python
#export visualisation
bar_2.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'bar_loyalty_status.png'))
```

```python
#create a smaller subset of the data by only including the necessary columns
df_4 = small[['order_hour_of_day','prices']]
```

```python
#create the line chart based on the subset data
line_2 = sns.lineplot(data = df_4, x = 'order_hour_of_day',y = 'prices', color ='#900C3F')
plt.title('Expenditure by Hour of the Day')
```

```python
#creating a scatterplot to determine the age and family situation
scatter = sns.scatterplot(x = 'Age', y = 'no_of_dependants',data = small, color='#900C3F')
plt.title('Relationship between Age and Number of Dependants')
```

The plot shows horizontal lines meaning that there is no relationship between the two. In other words the customer's age has no influence on the number of dependants they have.

```python
#export visualisation
scatter.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'scatter_age_no_dep.png'))
```

```python
#creating a scatterplot to determine the relationship between age and income
scatter_2 = sns.scatterplot(x = 'Age', y = 'income',data = small, color = '#900C3F')
plt.title('Relationship between Age and Income')
```

According to the chart we can see that most customers earn $200.000 and below. However, there are many customers above 40 who earn more than $400.000 whereas there are no customers below 40 that earn more than this.

```python
#export visualisation
scatter_2.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'scatter_age_income.png'))
```
