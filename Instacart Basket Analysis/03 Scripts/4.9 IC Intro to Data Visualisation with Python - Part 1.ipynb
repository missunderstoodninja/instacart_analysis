```python
#update column names
customers = customers.rename(columns={'First Name': 'first_name'})
customers = customers.rename(columns={'Surnam': 'surname'})
customers = customers.rename(columns={'STATE': 'state'})
customers = customers.rename(columns={'n_dependants': 'no_of_dependants'})
customers = customers.rename(columns={'fam_status': 'marital_status'})
```
