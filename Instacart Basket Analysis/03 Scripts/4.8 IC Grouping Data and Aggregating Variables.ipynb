```python
#Loyalty
```

```python
#Split the data and create column for maximum orders
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(100)
```

```python
pd.options.display.max_rows = None
```

```python
#assigning the loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
value_counts = ords_prods_merge['user_id'].value_counts()
unique_count = len(value_counts)
print(unique_count)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
#check the average spend between the loyalty groups
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

New Customers seem to have a higher price average then regular and loyal customers. Its surprising to see that loyal customers have a lower price average.

```python
#creating a spender flag for each customer
ords_prods_merge['average_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(20)
```

```python
#create the spender type flag
ords_prods_merge.loc[ords_prods_merge['average_spend'] < 10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spend'] >= 10, 'spender_flag'] = 'High spender'
```

```python
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
#check data
ords_prods_merge[['user_id', 'average_spend', 'spender_flag']].head(20)
```

```python
#creating a frequent buyer flag
```

```python
#Split the data and create column for median days between orders
ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head(20)
```

```python
#create the buying frequency flag
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] > 20, 'frequent_buyer_flag'] = 'Non-frequent_customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] <= 20) & (ords_prods_merge['median_days_between_orders'] > 10), 'frequent_buyer_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] <= 10, 'frequent_buyer_flag'] = 'Frequent_customer'
```

```python
#check data
ords_prods_merge[['user_id', 'median_days_between_orders', 'frequent_buyer_flag']].head(20)
```

```python
#turn the values to missing
ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```
