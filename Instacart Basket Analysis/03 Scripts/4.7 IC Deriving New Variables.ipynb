### Data Wrangling

#### Prices

```python
#create subset of the data
df = df_ords_prods_merge[:1000000]
```

```python
#define the new function using an if statement
def price_label(row):
  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df_ords_prods_merge.shape
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
#validate the value counts
df['prices'].max()
```

```python
#use the .loc method to create the if statement
```

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
#apply the new column to the full data
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

#### Day of the Week

```python
#get frequency of the days_of_the_week
df_ords_prods_merge['order_day_of_week'].value_counts(dropna = False)
```

```python
#determine which day of the week is busy using a for-loop and create a list 
result = []

for value in df_ords_prods_merge["order_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#create a new column to enter the results list in
df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

```python
#create a new column for busiest days
#determine which day of the week is busy using a for-loop and create a list 
result2 = []

for value in df_ords_prods_merge["order_day_of_week"]:
  if value == 0 or value == 1:
    result2.append("Busiest day")
  elif value == 4 or value == 3:
    result2.append("Least busy")
  else:
    result2.append("Regularly busy")
```

```python
result2
```

```python
#create a new column to enter the results list in
df_ords_prods_merge['busiest_day'] = result2
```

```python
print(df_ords_prods_merge[['order_day_of_week','busiest_day']])
```

#The days are now coded correctly with days 2,5 and 6 now regularly busy and 3 & 4 now least busy

```python
#rename column name
df_ords_prods_merge = df_ords_prods_merge.rename(columns={'busiest_day':'busiest_days'})
```

#### Hour of the day

```python
#identify the busiest hours of the day
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
#save results to a list
result3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
    result3.append("Most orders")
  elif value == 17 or value == 8 or value == 18 or value == 19 or value == 20 or value == 7 or value == 21 or value == 22:
    result3.append("Average orders")
  else:
    result3.append("Fewest orders")
```

```python
result3
```

```python
#create a new column to enter the results list in
df_ords_prods_merge['busiest_period_of_day'] = result3
```

```python
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
